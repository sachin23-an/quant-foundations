```python
prices = [100, 102, 101, 105, 110]

print(prices)
```

```python
print("Max price:", max(prices))
print("Min price:", min(prices))
```

```python
returns = []

for i in range(1, len(prices)):
    r = (prices[i] - prices[i-1]) / prices[i-1]
    returns.append(round(r, 4))

print(returns)
```

```python
print([r * 100 for r in returns])
```
