```python
df = pd.read_csv("../data/bhavcopy.csv")

df.head()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip() # it removes the leading and trailing spaces from the column names
```

```python
# extracting closing prices as numpy array
close = df["CLOSE_PRICE"].values

print(type(close))  # should show numpy array
```

```python
import numpy as np

prices = df['CLOSE_PRICE'].values # it extracts the 'CLOSE_PRICE' column from the DataFrame and converts it into a numpy array. This allows us to perform efficient numerical operations on the closing prices, such as calculating returns, volatility, etc.

returns = (prices[1:] - prices[:-1]) / prices[:-1]

df['Returns'] = np.nan
df.loc[1:, 'Returns'] = returns # it assigns the calculated returns to a new column 'Returns' in the DataFrame. The first row is set to NaN because there is no previous price to calculate the return for the first day.
Returns = df['Returns'].values
```

```python
Returns
```

```python
df.head(1)
```

```python
log_returns = np.log(prices[1:] / prices[:-1]) # it calculates the log returns by taking the natural logarithm of the ratio of consecutive prices. This is often used in finance because it has nice properties, such as being time-additive and more normally distributed than simple returns.

df['Log_Returns'] = np.nan
df.loc[1:, 'Log_Returns'] = log_returns
Log_Returns = df['Log_Returns'].values
```

```python
df['Intraday_Range'] = (df['HIGH_PRICE'] - df['LOW_PRICE']) / df['LOW_PRICE'] # it calculates the intraday range as the difference between the high and low prices divided by the low price, and stores it in a new column 'Intraday_Range' in the DataFrame. This gives us a measure of the volatility within a single trading day.
```

```python
avg_volume = np.mean(df['TTL_TRD_QNTY']) # it calculates the average trading quantity (volume) by taking the mean of the 'TTL_TRD_QNTY' column in the DataFrame. This gives us an idea of the typical trading volume for the stocks in our dataset.

df['Volume_Spike'] = df['TTL_TRD_QNTY'] > 2 * avg_volume # it creates a new column 'Volume_Spike' in the DataFrame, which is True if the trading quantity for that row is greater than twice the average volume, and False otherwise. This can help us identify days with unusually high trading activity.
```

```python
df['Turnover_per_Trade'] = df['TURNOVER_LACS'] / df['NO_OF_TRADES'] # it calculates the turnover per trade by dividing the total turnover (in lacs) by the number of trades, and stores it in a new column 'Turnover_per_Trade' in the DataFrame. This gives us an idea of the average value of each trade.
```

```python
df['Bullish_Signal'] = (
    (df['Returns'] > 0) &
    (df['Volume_Spike'])
)
```

```python
print(np.mean(df['Returns']))
print(np.std(df['Returns']))
print(np.max(df['Returns']))
print(np.min(df['Returns']))
```
