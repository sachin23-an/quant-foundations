```python
prices = np.array([100, 102, 101, 105])

returns = (prices[1:] - prices[:-1]) / prices[:-1]
print(returns)
```

```python
log_returns = np.log(prices[1:] / prices[:-1])
print(log_returns) # log return is the natural logarithm of the ratio of consecutive prices. It is often used in finance because it has nice properties, such as being time-additive and more normally distributed than simple returns.
```

```python
np.std(returns) # volatility of returns
```
