```python
df[df["TTL_TRD_QNTY"] > 1000000].head(3) # it gives us all the rows where the total traded quantity is greater than 10 lakhs
```

```python
df.sort_values(by="CLOSE_PRICE", ascending=False).head() # it gives us the top 5 rows where the closing price is the highest
```

```python
df[df["SERIES"] == "EQ"] # it gives us all the rows where the series is EQ
```

```python
df["Range"] = df["HIGH_PRICE"] - df["LOW_PRICE"] # it creates a new column called Range which is the difference between the high and low price of the day
```
